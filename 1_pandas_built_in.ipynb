## CUSTOM PLOTS

### *Custom line*

```python
df2['a'].plot.line(ls='--', color='red', lw=4)
# '-'   >> continuous line (default)
# '--'  >> dashed line
# '-.'  >> dashed dot line
# ':'   >> dott line
```

### *Custom axis label & title*

```python
ax = df2['a'].plot.line()
ax.set(xlabel='X_Label',
       ylabel='Y_Label',
       title='TITLE')
```

### *Custom Legend Position*

```python
ax = df2.plot.line()
ax.legend(loc=0,  # lokasi legend yang diinginkan
          bbox_to_anchor=(1.0, 1.0)) # pergeseran legend dari lokasi awal
```
